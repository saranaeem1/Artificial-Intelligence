# KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

X = data[['x', 'y']].values
y = data['class'].values

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
```

```python
y_pred = knn.predict(X)
```

```python
print(classification_report(y_pred, y))
```
